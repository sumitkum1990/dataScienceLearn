```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c='r', alpha = .9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
